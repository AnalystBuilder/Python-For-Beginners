# Let's put it all together


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
import os
import time
```

```python

def automated_crypto_pull():
    url = 'https://coinmarketcap.com/currencies/bitcoin/'

    page = requests.get(url)

    soup = BeautifulSoup(page.text, 'html')

    crypto_name = soup.find('span', class_ = 'sc-1d5226ca-1 fLa-dNu').text

    crypto_price = soup.find('div', class_ = 'priceValue').text

    final_price = crypto_price.replace('$','')

    date_time = datetime.now()

    dict = {'Crypto Name':crypto_name,
             'Price':final_price,
             'TimeStamp': date_time}

    df = pd.DataFrame([dict])

    if os.path.exists(r'C:\Users\alexf\OneDrive\Documents\Python Tutorials\Crypto Web Puller\Crypto_Automated_Pull.csv'):
        df.to_csv(r'C:\Users\alexf\OneDrive\Documents\Python Tutorials\Crypto Web Puller\Crypto_Automated_Pull.csv', mode='a', header = False, index = False)
    else:
        df.to_csv(r'C:\Users\alexf\OneDrive\Documents\Python Tutorials\Crypto Web Puller\Crypto_Automated_Pull.csv', index = False)
    print(df)
```

```python
while True:
    automated_crypto_pull()
    time.sleep(3600)
```
