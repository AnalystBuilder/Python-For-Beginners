# Slicing Indexed Variables

```python
my_string = 'Happy Monday!'
             0123456789
```

```python
print(my_string[4])
```

```python
print(my_string[45])
```

```python
my_string = 'Happy Monday!'
             0123456789
```

```python
print(my_string[0:5])
```

```python
print(my_string[0:9:5])
```

```python
my_string = 'Happy Monday!'
             0123456789
                      -2-1
```

```python
print(my_string[-1])
```

```python
print(my_string[-7:])
```

```python
my_list = [1,2,3,'this is a list']
```

```python
print(my_list[3][0:5])
```
