# BeautifulSoup and Requests

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/allinone/phones'

requests.get(url)
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
print(soup.prettify())
```
