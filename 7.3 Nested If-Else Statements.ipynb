# Nested If/Else Statements

```python
if (25 < 10) or (1 < 3):
    if 25 < 10:
        print('It was this one')
    elif 1 < 3:
        print('It was the second one')
```

```python
grade = 100

if grade < 70:
    print(grade, '= F')
elif grade <= 74:
    print(grade, '= D')
elif grade <= 84:
    if grade < 77:
        print(grade, '= C-')
    elif grade <= 81:
        print(grade, '= C')
    elif grade <= 84:
        print(grade, '= C+')
    else:
        print('Your grade is within a C')
else:
    print('You scored higher than a C.')
    
```
